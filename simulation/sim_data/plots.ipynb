```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# First find max in the xschem sine wave
peaks, _ = find_peaks(vout01['v(vout+)-v(vout-)'], distance=20)
print(peaks)
peak_loc1 = 4806
print('Amplitude at peak_loc1 =', vout01['v(vout+)-v(vout-)'].loc[peak_loc1])

# Set the frequency and amplitude of the sine wave
frequency = 6.25e6
amplitude = vout01['v(vout+)-v(vout-)'].loc[peak_loc1]

# Set the sampling rate and duration of the waveform
sampling_rate = 1.1208e10
duration = 1e-6

# Generate the time array
time = np.arange(2.235e-7, duration, 1/sampling_rate)
shifted_time = time+2.5e-7

# Generate the sine wave
sine_wave = amplitude * np.sin(2*np.pi*frequency*shifted_time)

# Plot the sine wave
fig, ax = plt.subplots()
vout01.loc[2500:].plot(x = 'time', y='v(vout+)-v(vout-)', ax = ax, title = 'Comparison to ideal sine wave')
plt.xlabel('Time')
plt.ylabel('Vout (V)')


plt.plot(time, sine_wave)
plt.show()
plt.savefig("idea_sine_wave_comparison.svg", format="svg")
```

```python
# Calculate standard error of xschem generated sine wave
clean_sine_len = len(vout1['time'])-2500
se_generated = np.var(vout1['v(Vout+)-v(Vout-)'])/np.sqrt(clean_sine_len)
print(se_generated)


# Calculate standard error of ideal sine wave
idea_sine_len = len(sine_wave)
se_ideal = np.var(sine_wave)/np.sqrt(idea_sine_len)
print(se_ideal)
```
