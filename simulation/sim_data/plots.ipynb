### Shift Register Plots

#### Transient Simulations

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# First find max in the xschem sine wave
peaks, _ = find_peaks(vout1['v(vout+)-v(vout-)'], distance=20)
print(peaks)
peak_loc1 = 4806
print('Amplitude at peak_loc1 =', vout1['v(vout+)-v(vout-)'].loc[peak_loc1])

# Set the frequency and amplitude of the sine wave
frequency = 6.25e6
amplitude = vout1['v(vout+)-v(vout-)'].loc[peak_loc1]

# Set the sampling rate and duration of the waveform
sampling_rate = 1.1208e10
duration = 1e-6

# Generate the time array
time = np.arange(2.235e-7, duration, 1/sampling_rate)
shifted_time = time+2.5e-7

# Generate the sine wave
sine_wave = amplitude * np.sin(2*np.pi*frequency*shifted_time)

# Plot ideal sine wave
fig, ax = plt.subplots()
plt.plot(time, sine_wave, '--')

# Plot the sine wave
vout1.loc[2500:].plot(x = 'time', y='v(vout+)-v(vout-)', ax = ax, title = 'Comparison Between Ideal and Generated Sine Waves')
plt.xlabel('Time')
plt.ylabel('Vout (V)')
plt.savefig("idea_sine_wave_comparison.svg", format="svg")
```

```python
fig, ax = plt.subplots()
start = 0
vout1.loc[start:].plot(x = 'time', y='v(vout+)-v(vout-)', ax = ax, title = 'Transient Simulation of the Sine Wave Generator')
plt.xlabel('Time')
plt.ylabel('Vout (V)')
plt.savefig("sin_vout_tran.svg", format="svg")
```
